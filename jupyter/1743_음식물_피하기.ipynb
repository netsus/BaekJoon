```python
! # 시작시간 체크
```

    입력 조건
    통로의 세로 길이: 1 ≤ N ≤ 100
    통로의 가로 길이: 1 ≤ M ≤ 100
    음식물 쓰레기(#)의 개수: 1 ≤ K ≤ NxM

```python
! # 끝 시간 체크
```

### 풀이

```python
import sys
sys.setrecursionlimit(10**7)
N, M, K = map(int,input().split())
food = [[*map(int,input().split())] for _ in range(K)]

g = [[0]*M for _ in range(N)]
visited = [[0]*M for _ in range(N)]

for f in food:
    r,c=f
    g[r-1][c-1]=1
    
result=cnt=0

dr=[1,-1,0,0]
dc=[0,0,1,-1]

def dfs(r,c):
    global cnt
    visited[r][c]=1
    for r_,c_ in zip(dr,dc):
        nr=r+r_; nc=c+c_
        if 0<=nr<N and 0<=nc<M and visited[nr][nc]==0 and g[nr][nc]==1:
            visited[nr][nc]=1
            cnt+=1
            dfs(nr,nc)

for r in range(N):
    for c in range(M):
        if visited[r][c]==0 and g[r][c]==1:
            cnt=1
            dfs(r,c)
            result=max(result,cnt)

print(result)
```
