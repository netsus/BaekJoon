수식은 0~9까지의 정수와 연산자 +, -, x 로 이루어져있다. 정수로 시작하고, 정수와 연산자가 번갈아서 나옵니다. <br>
항상 올바른 수식만 주어지기 때문에 주어지는 N은 항상 홀수입니다. (숫자로 시작, 숫자로 끝)

상상코딩: 연산자는 최소1개 ~ 최대 9개이다. 연산자를 중심으로 Combination하여 괄호를 1개 치면 될거같다. 연속된 두 연산자가 한번에 있으면 안된다는 조건으로 모든 Combination을 어떻게 구할 수 있을까?

### 제출풀이

```python
from itertools import combinations
def non_consecutive_combinations(lst, r):
    return [tuple(lst[j+i] for i, j in enumerate(combi)) for combi in combinations(range(len(lst)-r+1), r)]
def cal(sik):
    for i in range(len(sik)//2):
        a=sik.pop(0)
        b=sik.pop(0)
        c=sik.pop(0)
        sik.insert(0,str(eval(a+b+c)))
    return int(sik[0])
N=int(input())
sik=list(input())
if N==1:
    print(sik[0])
else:
    num_op=N//2
    k=int(num_op/2+0.5) # 반올림 -> 최대괄호개수
    result=[]
    op_idx=[i for i in range(1,N+1,2)]
    for cnt in range(1,k+1):
        op_li=non_consecutive_combinations(op_idx,cnt)
        for op in op_li:
            r=[];flag=0
            for i,e in enumerate(sik):
                if flag:
                    a=r.pop()
                    r.append(str(eval(f"{a}{opr}{e}")))
                    flag=0
                elif i in op:
                    flag=1
                    opr=e
                else:
                    r+=e
                    flag=0
            result.append(cal(r))
    print(max(result))
```

## 16439 치킨치킨치킨
여러 종류의 치킨을 입력받고 여기서 3개만 시켜 모든 사람들의 만족도를 최대한으로 내는 문제입니다.

    회원수 N, 치킨 종류 M
    1번회원은 5번치킨이 가장 만족도 높다
    1, 3, 5번치킨 시키면 1번회원 5점, 2번 5점, 3번 3점해서 총 13점으로 만족도 최대된다.(한사람당 만족도는 최대값 1개)
    회원의 수 N (1 ≤ N ≤ 30) 과 치킨 종류의 수 M (3 ≤ M ≤ 30)
    
https://www.acmicpc.net/problem/16439

### 복습 1회차

#### 제출 풀이

```python
from itertools import combinations
N,M=map(int,input().split())
l=[]
for i in range(N):
    l.append([*map(int,input().split())])
r=0
for a,b,c in combinations(list(range(M)),3):
    manjok=sum(max(p[a],p[b],p[c]) for p in l)
    r=max(r,manjok)
print(r)
```

### 상상코딩
- Combination으로 3개의 열을 고르고 -> 각 행에서 최대값 더한거 모두 비교

        3개의 치킨을 골라야 하기 때문에, 각 행에대해 최대값만 더할 수 없다.
        3개의 열을 고르고, 그중 각 행에 대해 최대값 끼리 합한것이 최대가 되어야 한다.    

### 제출 풀이

```python
from itertools import combinations
N, M = map(int,input().split())
chik = [list(map(int,input().split())) for i in range(N)]
result=0
for a, b, c in combinations(range(M), 3):
    result=max(result,sum([max(r[a],r[b],r[c]) for r in chik]))
print(result)
```

## 2961 도영이가 만든 맛있는 음식

    지금 도영이의 앞에는 재료가 N개N(1 ≤ N ≤ 10) 있다. 도영이는 각 재료의 신맛 S와 쓴맛 B를 알고 있다. 
    여러 재료를 이용해서 요리할 때, 그 음식의 신맛은 사용한 재료의 신맛의 곱이고, 쓴맛은 합이다.
    도영이는 재료를 적절히 섞어서 요리의 신맛과 쓴맛의 차이를 작게 만들려고 한다.
    재료는 적어도 하나 사용해야 한다.
   **재료의 신맛과 쓴맛이 주어졌을 때, 신맛과 쓴맛의 차이가 가장 작은 요리의, 그 차이를 출력**

    4
    1 7
    2 6
    3 8
    4 9
    -> 1
**2, 3, 4번 재료를 사용한다면, 요리의 신맛은 2×3×4=24, 쓴맛은 6+8+9=23이 된다. 차이는 1이다.**

https://www.acmicpc.net/problem/2961

### 복습 1회차

```python
재료는 적어도 하나, 신맛은 곱, 쓴맛은 합
```

```python
from itertools import combinations
from functools import reduce
N=int(input())
l=[[*map(int,input().split())] for _ in range(N)]
m=lambda x,y:x*y
r=99**9
for i in range(1,N+1):
    for c_li in combinations(l,i):
        sin=reduce(m,(c[0] for c in c_li))
        ssun=sum(c[1] for c in c_li)
        r=min(r,abs(sin-ssun))
print(r)
```

## 2503 숫자 야구
https://www.acmicpc.net/problem/2503

### 복습 1회차

```python
from itertools import permutations
l=[input().split() for _ in range(int(input()))]
al=[i for i in permutations(list(range(1,10)),3)]
def base(cad,ans):
    s=b=0
    for c,a in zip(cad,ans):
        if c==a:
            s+=1
        elif c in ans:
            b+=1
    return s,b
r=[]
for c,s,b in l:
    s=int(s);b=int(b)
    c=[*map(int,c)]
    for a in al:
        if base(c,a)==(s,b):
            r.append(a)
    al=r[:];r=[]
print(len(al))
```

### 제출 풀이

```python
from itertools import permutations
li = [''.join(map(str,x)) for x in permutations(range(1,10),3)]
for _ in range(int(input())):
    r,s,b=map(int,input().split());r=str(r)
    for num in li[::-1]:
        S=0;B=0
        for i in range(3):
            if num[i] == r[i]:
                S+=1
            elif num[i] in r:
                B+=1
        if (S,B)!=(s,b):
            li.remove(num)
print(len(li))
```

#### permutation을 이용한 오래걸리는 풀이 - 2204ms

```python
from itertools import permutations
N=int(input())
n=[*map(int,input().split())]
op=[]
for o,k in zip(['+','-','*','//'],[*map(int,input().split())]):
    op.extend([o]*k)
result=[]
M=-(99**9);m=(99**9)
for l in list(set(permutations(op,N-1))):
    r=n[0]
    for i in range(N-1):
        if not(l[i]=='//' and n[i+1]==0):
            if l[i]=='//' and int(r)<0:
                r=str(eval(f"-({int(r)*-1}{l[i]}{n[i+1]})"))
            else:
                r=str(eval(f"{r}{l[i]}{n[i+1]}"))
    r=int(r)
    result.append(r)
    M=max(M,r)
    m=min(m,r)
print(f"{M}\n{m}")
```

```python
from itertools import combinations
def non_consecutive_combinations(lst, r):
    return [tuple(lst[j+i] for i, j in enumerate(combi)) for combi in combinations(range(len(lst)-r+1), r)]
def cal(sik):
    return eval(''.join(sik))
N=int(input())
sik=list(input())
if N==1:
    print(sik[0])
else:
    num_op=N//2
    k=int(num_op/2+0.5) # 반올림 -> 최대괄호개수
    result=[]
    op_idx=[i for i in range(1,N+1,2)]
    for cnt in range(1,k+1):
        op_li=non_consecutive_combinations(op_idx,cnt)
        for op in op_li:
            r=[];flag=0
            for i,e in enumerate(sik):
                if flag:
                    a=r.pop()
                    r.append(str(eval(f"{a}{opr}{e}")))
                    flag=0
                elif i in op:
                    flag=1
                    opr=e
                else:
                    r+=e
                    flag=0
            result.append(cal(r))
    print(max(result))
```

### 제출 풀이

```python
N=int(input())
H=[[*map(int,input().split())] for _ in range(N)]
dx=[1,-1,0,0];dy=[0,0,1,-1];C=[]
b=[(i,j) for i in [-1,0,1] for j in [-1,0,1]]
b+=[(2, 0), (-2, 0), (0, 2), (0, -2)]
for i in range(1,N-1):
    row=[]
    for j in range(1,N-1):
        cost=H[i][j]
        for x,y in zip(dx,dy):
            cost+=H[i+x][j+y]
        row.append(cost)
    C.append(row)
M=len(C)
visited=[[0]*M for _ in range(M)]
def dfs(start, cost, seed):
    global result
    if seed==3:
        result=min(result,cost)
        return
    for i in range(start,M):
        for j in range(0,M):
            B = [visited[i+x][j+y] for x,y in b if 0<=i+x<M and 0<=j+y<M]
            if visited[i][j]+sum(B)==0:
                visited[i][j]=1
                dfs(i, cost+C[i][j], seed+1)
                visited[i][j]=0
result=9**99
dfs(0,0,0)
print(result)
```
