### 문제

상상 코딩

    주어진 문자열에서 PPAP가 지워지는 모든 경우에 대해 큐에 넣어가며, BFS를 돌아 처음으로 P가 나오면 PPAP를 출력하고,
    다 돌았는데도 P가 나오지 않으면 NP를 출력하는 방식? -> 시간 초과

실제 풀이

    주어진 문자열에 대해, 
        하나 씩 stack에 넣고
            stack의 마지막이 PPAP이면 이를 P로 바꾼다
    최종 스택에 P가 남아있다면 PPAP, 아니면 NP

핵심 아이디어

    PPAP를 경우의 수를 따지지 않고, 모두 P로 바꿔도 
        결과가 P가 된다면 항상 PPAP가 성립되며
        결과가 P가 안된다면 항상 NP가 성립된다.       
    

```python
! # 끝 시간 체크
```

### 풀이

```python
stack=[]
for w in input():
    stack += w
    if stack[-4:]==[*'PPAP']:
        stack[-3:]=[]
print('PPAP' if stack==['P'] else 'NP')
```

### 연습

```python
from collections import deque
w=input()
q=deque([w])
while q:
    word = q.popleft()
    if "PPAP" in word:
        q.append(word.replace("PPAP","P",1))
    elif word=="P":
        print("PPAP")
        break
if word!="P":
    print("NP")
```
