### 풀이

```python
C,R=map(int,input().split())
k=int(input())
g=[[0 for _ in range(C)] for _ in range(R)]

if k > C*R:
    print(0)
    exit()

direct=[(1,0),(0,1),(-1,0),(0,-1)]
chg=r=c=0
g[0][0] = 1 # 출발점

for i in range(2,k+1):
    while 1:
        nr = r + direct[chg][0]
        nc = c + direct[chg][1]
        if 0<=nr<R and 0<=nc<C and g[nr][nc]==0:
            r=nr;c=nc
            g[r][c]=i
            break
        else:
            chg = (chg+1)%4

print(c+1,r+1)
```
