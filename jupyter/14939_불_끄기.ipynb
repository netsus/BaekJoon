상상 코딩
    
    불 켜진곳은 1, 꺼진곳은 0 2진수 10개를 갖는 리스트로 표현한다.
    브루트포스로 반복하며, 가장자리 일때 고려해서 십자가 모양을 본다. 모두 1인 경우 0으로 바꿔준다.
    모두 1이 아닐 때는 최대한 겹칠 때, 해당 인덱스와 몇개 겹쳤는지를 최대힙에 저장한다.
    최대힙 없으면, 해당 개수 리턴
    최대힙 있으면, 최대한 겹쳤을 때 1을 0으로 바꿔가며, 모두 0인지 검사. 최대힙에서 뺴낸건 used_list에 저장
    다음 for문에 반복하며 used_list에 존재하는걸 또 최대힙에 넣어야되는 경우가 있으면 (반복되는 위치가 있다면, -1 출력하고 끝)    

### 풀이

```python
from copy import deepcopy
arr0 = [list(input()) for _ in range(2)]
dic={"O":"#","#":"O"}
res=100
for k in range(1024):
    arr=deepcopy(arr0); cnt=0
    for j in range(10):
        if k&(1<<j):
            cnt+=1
            for r,c in (0,j-1),(0,j),(0,j+1),(1,j):
                if 0<=c<10:
                    arr[r][c] = dic[arr[r][c]]
    for i in range(9):
        for j in range(10):
            if arr[i][j]=="O":
                cnt+=1
                for r,c in (i,j),(i+1,j-1),(i+1,j),(i+1,j+1),(i+2,j):
                    if r<10 and 0<=c<10:
                        arr[r][c] = dic[arr[r][c]]
    if all(i=='#' for i in arr[-1]):
        res=min(res,cnt)
print(res if res<101 else -1)
```

### 주석 달린 풀이

```python
from copy import deepcopy
arr0 = [list(input()) for _ in range(10)] # 전구배열 입력
dic = {'O':'#','#':'O'} # 스위치 켜고, 끄고 딕셔너리
res = 101 # 최종 출력값 초기화
for k in range(1024): # 첫 줄의 모든 경우수
    arr = deepcopy(arr0); cnt = 0 # 스위치 누른 횟수
    for j in range(10): # 첫줄에 10개 전구 하나씩 탐색
        if k&(1<<j): # j번째 스위치를 누른 경우(눌려있는 경우)
            cnt += 1 # 스위치 누르기
            for r,c in (0,j-1),(0,j),(0,j+1),(1,j): # 맨 윗줄 4방향 보기
                if 0<=c<10: # 열이 삐져나가지 않은 경우
                    arr[r][c] = dic[arr[r][c]] # 스위치 딸깍!
    for i in range(9): # 첫줄부터 9번쨰
        for j in range(10): # 10개 열
            if arr[i][j]=='O':
                cnt += 1
                for r,c in (i,j),(i+1,j-1),(i+1,j),(i+1,j+1),(i+2,j):
                    if r<10 and 0<=c<10: 
                        arr[r][c] = dic[arr[r][c]]
    if all(c=='#' for c in arr[-1]): # 막줄에 모두 꺼진 경우
        res = min(res,cnt)
print(res if res<101 else -1)
```
