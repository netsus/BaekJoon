### 풀이

```python
## pypy3로 제출해야 한다.
import sys
input=sys.stdin.readline
N=int(input())
mat = [list(map(int,input().split())) for _ in range(N)]
ans=1e9

for k in range(2**N):
    m = [i[:] for i in mat]
    cnt=0
    for j in range(N):
        if k&(1<<j):
            cnt+=1
            for r,c in (0,j-1),(0,j),(0,j+1),(1,j):
                if 0<=c<N:
                    m[r][c] = not m[r][c]
    for i in range(N-1):
        for j in range(N):
            if m[i][j]:
                cnt+=1
                for r,c in (i,j),(i+1,j-1),(i+1,j),(i+1,j+1),(i+2,j):
                    if r<N and 0<=c<N:
                        m[r][c] = not m[r][c]
    if not any(m[-1]):
        ans=min(ans,cnt)
print(ans if ans!=1e9 else -1 )
```
