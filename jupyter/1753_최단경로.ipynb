### 문제


```python
! # 시작시간 체크
```

문제 요약
    
    첫째 줄에 V(정점의 개수), E(간선의 개수)가 주어집니다. 둘째 줄에 시작정점의 번호 K가 주어집니다.
    셋째 줄부터 E개의 줄에 걸쳐 시작노드, 끝노드, 가중치가 주어집니다.
    V개 줄에 걸쳐 시작노드에서 각 노드별로 최단거리를 출력해야 합니다.

```python
! # 끝 시간 체크
```

### 풀이

```python
import heapq
V,E = map(int,input().split())
K = int(input())
dist = [1e9]*(V+1)
graph = [[] for _ in range(V+1)]
for _ in range(E):
    u,v,w = map(int,input().split())
    graph[u].append((w,v))
q = [(0,K)]
dist[K] = 0
while q:
    cur_w, cur_n = heapq.heappop(q)
    if dist[cur_n] < cur_w:
        continue
    for w, n in graph[cur_n]:
        nxt_w = cur_w + w
        if dist[n] > nxt_w:
            dist[n] = nxt_w
            heapq.heappush(q, (nxt_w,n))
for c in dist[1:]:
    print("INF" if c==1e9 else c)
```

### 연습

```python
import heapq

V,E = map(int,input().split())
K = int(input())
graph = [[] for _ in range(V+1)]
dist = [1e9] * (V+1)
q = []
for i in range(E):
    u, v, w = map(int,input().split())
    graph[u].append([w,v])

dist[K] = 0
heapq.heappush(q, [0, K])
while q:
    cur_w, cur_node = heapq.heappop(q)
    
    if dist[cur_node] < cur_w:
        continue
        
    for w, nxt_node in graph[cur_node]:
        nxt_w = cur_w + w
        if nxt_w < dist[nxt_node]:
            dist[nxt_node] = nxt_w
            heapq.heappush(q, [nxt_w, nxt_node])
for i in dist[1:]:
    print(i if i!=1e9 else "INF")
```
