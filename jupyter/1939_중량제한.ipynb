```python
! # 시작시간 체크
```

상상 코딩
    
    특정 섬에서 다른 섬으로 이동할때 갈 수 있는 경로 중에, 중량이 최대인 경우를 구하는 것이다.
    즉, 갈 수 있는 경로중에 다리의 최소값이 최대인 경로를 찾는 문제
    
    섬은 만개 이하, 다리는 10만개 이하
    다리의 가중치를 내림차순으로 정렬하고, 하나씩 추가하면서, 공장이 있는 두 섬이 연결되었는지를 검사하는 느낌이 떠오른다.
    어떻게 풀었는지 찾아보자

```python
! # 끝 시간 체크
```

### 풀이

```python
import sys
input = sys.stdin.readline
n,m = map(int,input().split())
graph = sorted([tuple(map(int, input().split())) for _ in range(m)], key=lambda x:x[2])[::-1]
start, end = map(int,input().split())

p=list(range(n+1))
def find(u):
    if u != p[u]:
        p[u] = find(p[u])
    return p[u]
def union(u, v):
    root1 = find(u)
    root2 = find(v)
    p[root2] = root1
res=0
for u, v, weight in graph:
    if find(start)!=find(end):
        union(u,v)
        res=weight
print(res)
```
