문제 요약

    처음에 정점의 개수 N이 주어집니다. 다음 N개의 줄 동안, i번쨰 줄에 j번째 숫자가 1인 경우 i에서 j로 가는 간선이 존재한다는 뜻입니다. 
    i번째 줄의 i번쨰 숫자는 항상 0입니다. 즉, 자기자신으로가는 간선은 없습니다. NxN행렬에 i에서 j로 가는 경로가 있으면 1, 없으면 0을 출력해야합니다.

상상코딩

    리스트에 인덱스가 노드, 값이 연결된 노드로 생성.
    그다음은 어떻게하지? 모르겠다.

```python
! # 끝 시간 체크
```

### 풀이

```python
N = int(input())
graph = [list(map(int,input().split())) for _ in range(N)]

for k in range(N):
    for i in range(N):
        for j in range(N):
            if graph[i][j] == 1 or (graph[i][k] == 1 and graph[k][j] == 1):
                graph[i][j] = 1
for r in graph:
    print(*r)
```
