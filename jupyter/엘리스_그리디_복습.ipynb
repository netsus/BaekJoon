## 20115_에너지 드링크

https://www.acmicpc.net/problem/20115

- 문제 요약
    
    1. 임의의 서로 다른 두 에너지 드링크를 고른다.
    2. 한쪽 에너지 드링크를 다른 쪽 에너지 드링크에 모두 붓는다. 단, 페인은 야근 후유증으로 인해 손이 떨려, 붓는 과정에서 원래 양의 절반을 바닥에 흘리게 된다.
    3. 다 붓고 남은 빈 에너지 드링크는 버린다.
    4. 1~3 과정을 에너지 드링크가 하나만 남을 때까지 반복한다.
    페인은 합쳐진 에너지 드링크의 양을 최대로 하려 한다. 불쌍한 페인을 도와주자!
    

- 상상 코딩

    
    가장 큰 드링크를 제외하고 나머지를 한 번 씩 쏟아서 최대를 만들어야 한다.
    전체를 더하고, 2로 나눈 것에 최대값 /2 한것을 더하면 된다.
    
왜 이게 그리디일까?

    각 반복에서 최선의 선택 즉, 제일 작은 드링크를 큰 드링크에 따르면 된다.
    즉, 제일 큰 드링크를 제외하고를 모두 따라서 합쳐야 한다.
    즉, 각 순간에 최선의 선택 -> 가장 작은 드링크를 가장 큰 드링크에 붓는 선택 -> 그리디

```python
input()
num_list = list(map(int,input().split()))
print((sum(num_list)+max(num_list))/2)
```

## 13305 주유소
https://www.acmicpc.net/problem/13305

- 문제 요약

    
    각 도시를 연결하는 도로의 길이를 입력으로 받아 제일 왼쪽 도시에서 제일 오른쪽 도시로 이동하는 최소의 비용을 계산하는 프로그램
    첫 번째 줄에는 도시의 개수를 나타내는 정수 N(2 ≤ N ≤ 100,000)
    다음 줄에는 인접한 두 도시를 연결하는 도로의 길이가 제일 왼쪽 도로부터 N-1개의 자연수로 주어진다. 
    다음 줄에는 주유소의 리터당 가격이 제일 왼쪽 도시부터 순서대로 N개의 자연수로 주어진다. 
    제일 왼쪽 도시부터 제일 오른쪽 도시까지의 거리는 1이상 1,000,000,000 이하의 자연수
    리터당 가격은 1 이상 1,000,000,000 이하의 자연수이다. 


상상 코딩

    각 도로를 반복하면서, 그 도로 왼쪽의 주유소 중에 최소값으로 그 도로를 건너면 된다. 
    도로를 건널 때 마다 주유소 최소 값을 갱신해가며 가면 된다.
    
그리디인 이유

    각 load에 대해, 가장 최소가격의 주유소 기름으로 가야한다.
    각 순간에 가장 최적의 선택을 한다는 점이 그리디
    각 길에서 그 앞에 있는 주유소중 가장 싼 주유소의 기름을 사용한다는 포인트
    

```python
N=4
loads=[2,3,1]
oils=[5,2,4,1]
loads=[3,3,4]
oils=[1,1,1,1]
oil=oils[0]
result=0
for i in range(len(loads)):
    oil=min(oil,oils[i])
    result+=oil*loads[i]
```

### 제출풀이

```python
N=input()
loads=list(map(int,input().split()))
oils=list(map(int,input().split()))
oil=oils[0]
result=0
for i in range(len(loads)):
    oil=min(oil,oils[i])
    result+=oil*loads[i]
print(result)    
```

## 2812 크게 만들기
https://www.acmicpc.net/problem/2812

- 문제 요약

    


- 상상 코딩


    K개의 숫자를 지웠을 때, 가장 커야한다. -> 제일 앞자리에 큰 숫자가 와야한다.
    맨 왼쪽부터 진행하며 작은 수를 지움으로써 맨 왼쪽에 가장 큰수가 오도록 해야한다.
    맨 앞부터 쭉 보며, 스택을 이용해서 맨 앞을 가장 큰수로 유지한다.
    즉, 뒷 숫자가 스택의 맨 위보다 크면 스택 팝하고 넣고, 스택의 맨 위 보다 작으면 그냥 넣고를 반복
    같은 숫자가 반복되는 경우, stack에서 제외가 안될 수 있기 때문에 tmp_k를 둬서 따로 빼주는 숫자개수를 세주고, 마지막엔 N-K개의 숫자를 앞에서부터 보여주면 된다.
    
- 그리디인 이유

    nums를 반복하며, 순간순간 뺄수와 빼지 않을 수를 구분하여 최적의 숫자만 남기기 때문이다.
    

```python
N,K = map(int,input().split())
nums = [*map(int,list(input()))]
tmp_k = K
stack = []
for n in nums:
    while stack and tmp_k and stack[-1]<n:
        stack.pop()
        tmp_k-=1
        if tmp_k==0:
            break
    stack.append(n)
result = "".join(map(str,stack))
print(result[:N-K])
```
