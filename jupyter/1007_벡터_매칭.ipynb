문제 요약
    첫 줄에 테스트케이스 T가 주어지고, 각 테스트 케이스에 대해 점의 개수 N이 주어집니다.
    다음 N개의 줄에 x,y 좌표가 주어집니다. N은 항상 짝수입니다.
    좌표 평면에 N개의 점이 있을 때, 모든 점을 사용해서 벡터를 N/2개 만들때 벡터의 합이 최소가될 때 최소값을 출력해야합니다.

첫번째 예시는 4개의 점에의해 정사각형이 만들어지기 때문에, 서로 반대방향의 벡터를 가로나 세로로 만들면 합했을 때 0이 나와 최소값을 만들 수 있고,
두번째 예시는 두점이기 때문에 두 점에의한 벡터의 길이가 곧 벡터의 합이 되어 길이를 출력하면 됩니다.

```python
((200_000)**2 + (200_000)**2)**(1/2)
```

상상 코딩

    N은 20개 이하이므로, 20개에 대해 2점을 이어 벡터의 Permutation을 모두 생성합니다.
    20P2개 -> 380개. 이중 포문으로 서로 점이 포함되지 않는 두 벡터에 대해 합이 최소값이 될 때마다 갱신

### 풀이

```python
import sys, itertools
input=sys.stdin.readline
T=int(input())
for _ in range(T):
    N=int(input())
    points = []
    total_x,total_y = 0,0

    for _ in range(N):
        x,y = map(int,input().split())
        total_x +=x ; total_y += y
        points.append((x,y))

    comb = list(itertools.combinations(points, N//2))
    ans=3e5
    
    for c in comb[:len(comb)//2]: # len(comb)는 항상 짝수
        x1,y1 = 0,0
        for x,y in c:
            x1 += x ; y1 += y
        x2,y2 = total_x-x1,total_y-y1
        
        hab_vector = ((x2-x1)**2 + (y2-y1)**2)**(0.5)
        ans=min(ans,hab_vector)
    print(ans)
```

### 주석 달린 풀이

```python
import sys, itertools
input=sys.stdin.readline
T=int(input())
for _ in range(T):
    N=int(input()) # 점의 개수
    points = [] # 좌표의 리스트
    total_x,total_y = 0,0

    for _ in range(N):
        x,y = map(int,input().split())
        total_x +=x ; total_y += y # 모든 x의 합과 y의 합 저장
        points.append((x,y))

    comb = list(itertools.combinations(points, N//2))
    ans=3e5
    
    for c in comb[:len(comb)//2]: # len(comb)는 항상 짝수
        x1,y1 = 0,0
        for x,y in c:
            x1 += x ; y1 += y
        x2,y2 = total_x-x1,total_y-y1 # x와 y를 x1,y1과 x2,y2 두 그룹으로 절반 나누기
        
        hab_vector = ((x2-x1)**2 + (y2-y1)**2)**(0.5) # 절반 나눈 두 그룹간의 합벡터
        ans=min(ans,hab_vector)
    print(ans)
```
