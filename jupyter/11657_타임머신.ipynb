```python
! # 시작시간 체크
```

문제 요약
    
    첫 줄에 N(도시 개수)와 M(버스 개수)가 주어집니다. 그 다음부터 M개의 줄 동안, 각 버스 별로, 시작 도시, 도착 도시, 버스 타고 이동하는데 걸리는 시간이 주어집니다. 
    1번 도시에서 출발해 각 도시별로 도착하는데 걸리는 최소시간을 출력해야 합니다.(만약 어떤 도시로 가는 과정에서 무한이 음수가 나오는 경우 -1을 출력합니다.)

상상 코딩
    
    

```python
! # 끝 시간 체크
```

### 풀이

```python
N,M = map(int,input().split())
dist = [1e9]*(N+1)
edges = [list(map(int,input().split())) for _ in range(M)] # start, end, cost

def bf(start):
    dist[start] = 0
    for n in range(N):
        for e in range(M):
            cur, nxt, cost = edges[e]
            if dist[cur] != 1e9 and dist[nxt] > dist[cur] + cost:
                dist[nxt] = dist[cur] + cost
                if n==N-1:
                    return True
    return False

neg_cycle = bf(1)
if neg_cycle:
    print(-1)
else:
    for i in range(2,N+1):
        print(-1 if dist[i] == 1e9 else dist[i])
```

### 주석 달린 풀이

```python
N,M = map(int,input().split())
dist = [1e9]*(N+1) # 1e9(10억)은 무한대를 의미
edges = [list(map(int,input().split())) for _ in range(M)] # start, end, cost로 간선 이중리스트 생성

def bf(start): # 벨만 포드 알고리즘
    dist[start] = 0 # 시작 노드에 대해 초기화
    for n in range(N): # 전체 노드를 반복하며,
        for e in range(M): # 각 노드마다 "모든 간선"을 확인
            cur, nxt, cost = edges[e]
            # 현재 간선을 거쳐서 다른 노드로 이동하는 거리가 더 짧은 경우
            if dist[cur] != 1e9 and dist[nxt] > dist[cur] + cost:
                dist[nxt] = dist[cur] + cost
                if n==N-1: # n 번째 노드에서도 값이 갱신된다면 음수 순환이 존재
                    return True
    return False

neg_cycle = bf(1) # 벨만 포드 알고리즘 수행
if neg_cycle: # 음수 순환이 존재하는 경우
    print(-1)
else:
    for i in range(2,N+1): # 2번 노드부터 출력
        print(-1 if dist[i] == 1e9 else dist[i]) # 연결된 노드가 아니면 -1, 연결된 노드면 비용 출력
```

### 연습

```python
n,m = map(int,input().split())
edges = []
dist = [1e9]*(n+1)

for _ in range(m):
    a, b, c = map(int, input().split())
    edges.append((a,b,c))

def bf(start):
    dist[start] = 0
    for i in range(n):
        for j in range(m):
            cur, nxt, cost = edges[j]
            if dist[cur] != 1e9 and dist[nxt] > dist[cur] + cost:
                dist[nxt] = dist[cur] + cost
                if i == n-1:
                    return True
    return False           
```
