```python
from heapq import heappush, heappop
import sys
def find(n):
    if n == parents[n]:
        return n
    else:
        parents[n]=find(parents[n])
        return parents[n]

def union(n1, n2):
    p1 = find(n1)
    p2 = find(n2)
    parents[p1]=p2


N=int(input())
edge=[]
for i in range(N):
    costs = list(map(int,input().split()))
    for j, cost in enumerate(costs):
        if cost!=0:
            heappush(edge,(cost,i,j))

parents=list(range(N))

connected_edges=0
ans=0

while connected_edges != N-1:
    c,a,b = heappop(edge)
    if find(a) != find(b):
        union(a,b)
        connected_edges+=1
        ans+=c

print(ans)
```

### 주석 달린 풀이

```python
from heapq import heappush, heappop
import sys

def find(n): 
    # 루트 부모 찾기
    # 찾는 과정에서 같은 라인에 있는 노드들은 한 루트 부모로 통일 시켜준다.
    if n == parents[n]:
        return n
    else:
        parents[n]=find(parents[n])
        return parents[n]

def union(n1, n2):
    # n1, n2의 부모를 p2로 통일 시켜주기. 
    p1 = find(n1)
    p2 = find(n2)
    parents[p1]=p2


N=int(input())
parents=list(range(N))
edge=[]
for i in range(N):
    # 입력받은 인접행렬 형태의 그래프를 코스트 순으로 정렬하여 cost, i, j 형태의 리스트로 변환
    costs = list(map(int,input().split()))
    for j, cost in enumerate(costs):
        if cost!=0:
            heappush(edge,(cost,i,j))
            
connected_edges=0
ans=0

while connected_edges != N-1:
    c,a,b = heappop(edge)
    if find(a) != find(b):
        union(a,b)
        connected_edges+=1
        ans+=c

print(ans)
```
