```python
bod = [list(map(int,input())) for _ in range(9)]

vst_row = [[0]*9 for _ in range(9)]
vst_col = [[0]*9 for _ in range(9)]
vst_box = [[0]*9 for _ in range(9)]

box = lambda i,j:i//3*3+j//3

for i in range(9):
    for j in range(9):
        if bod[i][j]:
            vst_row[i][bod[i][j]-1] = 1
            vst_col[j][bod[i][j]-1] = 1
            vst_box[box(i,j)][bod[i][j]-1] = 1

def dfs(i,j):
    if i==9: 
        return 1
    
    if bod[i][j]:
        return dfs(i+(j+1)//9, (j+1)%9)
    
    for c in range(9):
        if vst_row[i][c] or vst_col[j][c] or vst_box[box(i,j)][c]:
            continue
        bod[i][j] = c+1
        vst_row[i][c]=vst_col[j][c]=vst_box[box(i,j)][c]=1
        if dfs(i+(j+1)//9, (j+1)%9):
            return 1
        bod[i][j] = 0
        vst_row[i][c]=vst_col[j][c]=vst_box[box(i,j)][c]=0
    return 0

dfs(0,0)
for x in bod:
    print(*x,sep='')
```
