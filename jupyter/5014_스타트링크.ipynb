### 풀이

```python
from collections import deque
F,S,G,U,D=map(int,input().split())
visited=[0]*(F+1)
if S==G:
    print(0)
else:
    q=deque([S])
    while q:
        c=q.popleft()
        for n in (c+U,c-D):
            if 0<n<=F and visited[n]==0 and c!=n:
                visited[n]=visited[c]+1
                q.append(n)
    print(visited[G] if visited[G] else "use the stairs")
```

### visited에 버튼 횟수 저장하는 다른 풀이

```python
from collections import deque
F,S,G,U,D=map(int,input().split())
visited=[-1]*(F+1)
visited[S]=0
q=deque([S])
while q:
    c=q.popleft()
    if c==G:
        print(visited[G]);exit()
    for n in (c+U,c-D):
        if 0<n<=F and visited[n]==-1:
            visited[n]=visited[c]+1
            q.append(n)
print("use the stairs")
```
