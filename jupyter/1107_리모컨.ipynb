채널 N으로 이동하기 위해서 버튼을 최소 몇 번 눌러야하는지 구하는 프로그램

첫째 줄에 수빈이가 이동하려고 하는 채널 N (0 ≤ N ≤ 500,000)이 주어진다.  둘째 줄에는 고장난 버튼의 개수 M (0 ≤ M ≤ 10)이 주어진다. 고장난 버튼이 있는 경우에는 셋째 줄에는 고장난 버튼이 주어지며, 같은 버튼이 여러 번 주어지는 경우는 없다.

상상 코딩

    주어진 숫자를 x 라고 할 떄, x가 100보다 크면 100 ~ 2x-100 까지 눌러야하는 횟수와 함꼐 for문을 돌면서,
    해당 숫자가 현재 리모콘의 고장난 버튼을 제외하고 누를 수 있는 경우 누른 횟수를 answer에 추가하고
    for문이 끝난 뒤 answer에서 최소값 출력
    x가 100보다 작으면, 0~200까지 돌면서 위와 같은 방향으로 진행

### 풀이

```python
n = int(input())
m = int(input())
s = set(input().split()) if m else set()
ans = abs(100-n)
for i in range(int(1e6)):
    if not set(str(i)).intersection(s):
        ans = min(ans, len(str(i))+abs(n-i))
print(ans)
```

```python
n = int(input()) # 이동하려는 채널 번호
m = int(input()) # 고장난 버튼 개수
s = set(input().split()) if m else set()  # 고장난 버튼이 있으면 set에 저장
ans = abs(100-n)             # 100번부터 쭉 해당 채널까지 가는 횟수로 초기화
for i in range(int(1e6)):    # 100만번 브루트 포스
    if not set(str(i)).intersection(s): # 현재 채널 버튼중 하나라도 고장난 버튼에 포함되지 않으면
        ans = min(ans, len(str(i))+abs(n-i)) # 현재 채널에서 타겟 채널까지 버튼 개수로 min 갱신
print(ans)
```
