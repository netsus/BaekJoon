문제 설명: 미로가 주어졌을 때, (1, 1)에서 출발하여 (N, M)의 위치로 이동할 때 지나야 하는 최소의 칸 수를 구하는 프로그램을 작성<br>
미로에서 1은 이동할 수 있는 칸을 나타내고, 0은 이동할 수 없는 칸<br>
한 칸에서 다른 칸으로 이동할 때, 서로 인접한 칸으로만 이동(상,하,좌,우)<br>

    입력
    첫째 줄에 두 정수 N, M(2 ≤ N, M ≤ 100)이 주어진다. 다음 N개의 줄에는 M개의 정수로 미로가 주어진다. 각각의 수들은 붙어서 입력으로 주어진다.

    출력
    첫째 줄에 지나야 하는 최소의 칸 수를 출력한다. 항상 도착위치로 이동할 수 있는 경우만 입력으로 주어진다.

예시 풀이: 맨 왼쪽 위 (1,1)에서 맨 오른쪽 아래 (N,M)으로 가기 위해서는 아래로 쭉간뒤 오른쪽으로 갔다가 위로간뒤 오른쪽으로 갔다가 아래로간뒤 오른쪽으로 가면된다.

    상상코딩: 맨 왼쪽 위에서 갈 수 있는 곳(N,M내부의 1값)에 대해, 이미 방문한 곳은 또 방문하지 않으며 BFS를 수행한다. 그래서 최초로 N,M도달할 떄의 길이가 최단거리이다.
    방문한 곳을 또 방문하지 않는 이유: 앞에서 방문한 곳을 나중에 방문한다면, 앞에서 방문했을 떄의 길이가 더 짧을것이기 때문이다.

```python
q = deque([(0,0)])


while q:
    x,y = q.popleft()
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if not ((0<= nx <= M-1) and (0 <= ny <= N-1)):
            continue
        # 다음에 방문할 곳이 방문불가하거나 이미 방문한 경우
        if g[ny][nx]==0 or v[ny][nx]==1:
            continue
        # 다음에 방문할 곳이 방문하지않았고, 방문가능한 경우
        elif g[ny][nx]==1 and v[ny][nx]==0:
            v[ny][nx] = v[y][x]+1
            q.append((nx,ny))
```

### 제출 풀이

```python
from collections import deque

N,M = map(int,input().split())
g = [list(map(int,input())) for _ in range(N)]

dx=[0,0,1,-1]
dy=[1,-1,0,0]

v = [[0]*M for _ in range(N)]
v[0][0]=1

q = deque([(0,0)])

while q:
    x,y = q.popleft()
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if not ((0<= nx <= M-1) and (0 <= ny <= N-1)):
            continue
        # 다음에 방문할 곳이 방문불가하거나 이미 방문한 경우
        if g[ny][nx]==0 or v[ny][nx]==1:
            continue
        # 다음에 방문할 곳이 방문하지않았고, 방문가능한 경우
        elif g[ny][nx]==1 and v[ny][nx]==0:
            v[ny][nx] = v[y][x]+1
            q.append((nx,ny))

print(v[-1][-1])
```
