문제 요약

    이미 가지고 있는 랜선의 개수 K와 필요한 랜선의 개수 N이 첫줄에 주어집니다. 다음 K줄 동안 가진 랜선의 길이가 주어집니다.
    K개의 랜선을 잘라 N개의 랜선을 만들때 N개 랜선중 최소길이의 최대값을 출력해야 합니다.

상상 코딩

    1부터 랜선의 K개 랜선중 최대값까지 이분탐색을 하며, 각 타겟 랜선 길이에 대해
    정수 몫을 구해 몇개의 랜선을 만들 수 있는지 구합니다. 그 개수가 N이상일 때 최대 길이를 출력합니다.

```python
! # 끝 시간 체크
```

### 풀이

```python
import sys
# input=sys.stdin.readline
N,K = map(int,input().split())
n_li=[int(input()) for _ in range(N)]

s,e=1,max(n_li)
result=0

while s<=e:
    trg_len = (s+e)//2
    wire_cnt = sum([wire//trg_len for wire in n_li if wire>=trg_len])
    if wire_cnt>=K:
        result=trg_len
        s=trg_len+1
    else:
        e=trg_len-1

print(result)
```
