상상코딩

    dfs이용 백트래킹 문제.
    vstd에 방문한 문자를 넣고, 방문하지 않은 알파벳에 대해 DFS로 백트래킹하며 방문한다.
    더이상 방문이 안되면 함수는 종료되고, 가장 멀리갔을 때를 출력

### 풀이

```python
dr,dc = [-1,1,0,0],[0,0,-1,1]
vst = set([(0,0,bod[0][0])])
ans = 1

while vst:
    r, c, v = vst.pop()
    for i in range(4):
        nr, nc = r+dr[i], c+dc[i]
        if 0<=nr<R and 0<=nc<C and bod[nr][nc] not in v:
            vst.add((nr,nc,v+bod[nr][nc]))
            ans = max(ans, len(v)+1)
print(ans)
```

### 주석 달린 풀이

```python
dr,dc = [-1,1,0,0],[0,0,-1,1] # 상, 하, 좌, 우
vst = set([(0,0,bod[0][0])]) # 행, 열, 보드 알파벳
ans = 1 # 최종 출력값

while vst: # BFS
    r, c, v = vst.pop() # 랜덤 위치 pop
    for i in range(4): # 4 방향 모두 탐색
        nr, nc = r+dr[i], c+dc[i]
        if 0<=nr<R and 0<=nc<C and bod[nr][nc] not in v: # 보드 내, 방문하지 않은 알파벳
            vst.add((nr,nc,v+bod[nr][nc])) # 방문
            ans = max(ans, len(v)+1) # 방문 길이 최대값 갱신
print(ans)
```

```python
from collections import defaultdict
R,C=map(int,input().split())
bod=[list(input()) for _ in range(R)]
result=1
vstd=defaultdict(bool)
vstd[bod[0][0]]=True

def dfs(r,c,cnt):
    global result
    result=max(result,cnt)
    for i,j in [(0,-1),(0,1),(-1,0),(1,0)]:
        nr,nc = r+i,c+j
        if 0<=nr<R and 0<=nc<C and not vstd[bod[nr][nc]]:
            now=bod[nr][nc]
            vstd[now]=True
            dfs(nr,nc,cnt+1)
            vstd[now]=False
dfs(0,0,1)
print(result)
```
