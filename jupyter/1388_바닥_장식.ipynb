### 문제


```python
! # 시작시간 체크
```

    문제 요약: '-' 나 '|' 중에서 연속으로 이어진건 하나의 나무판자로 계산한다.
    방 바닥의 세로크기 N 과 가로크기 M이 주어진다.
    0<=N<=50 자연수
    0<=M<=50 자연수

```python
! # 끝 시간 체크
```

### 풀이

```python
N,M=map(int,input().split())
l=[list(input()) for _ in range(N)]
result=0

for r in range(N):
    pre='#'
    for c in range(M):
        if l[r][c]=='-' and l[r][c]!=pre:
            result+=1
        pre=l[r][c]
for c in range(M):
    pre='#'
    for r in range(N):
        if l[r][c]=='|' and l[r][c]!=pre:
            result+=1
        pre=l[r][c]

print(result)
```

### 주석 달린 풀이

```python
N,M=map(int,input().split())
l=[list(input()) for _ in range(N)]
result=0

for r in range(N): # 각 행에 대해서 - 판자 개수 세기
    pre='#'
    for c in range(M):
        if l[r][c]=='-' and l[r][c]!=pre:
            result+=1
        pre=l[r][c]
for c in range(M): # 각 열에 대해서 | 판자 개수 세기
    pre='#'
    for r in range(N):
        if l[r][c]=='|' and l[r][c]!=pre:
            result+=1
        pre=l[r][c]

print(result)
```
