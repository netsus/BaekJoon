상상 코딩

    커지다 특정 수를 기점으로 작아지는 가장 긴 부분 수열을 구해야 한다.
    N은 최대 1000
    1000 x 1000 배열을 탐색한다.
    1000개의 원소에 대해, 각 원소가 꼭대기 수(앞에는 오름차순, 뒤에는 내림차순)일때 최대 부분 수열을 구한다.
    즉, for문을 1000번 반복하는데, 각 for문에선 DP로 1000번만 연산하여 최대 부분 수열을 구하는 것이다.
    즉 최대 시간복잡도는 O(N^2)

```python
! # 끝 시간 체크
```

### 풀이

```python
from bisect import bisect_left
import sys
input = sys.stdin.readline
N = int(input())
arr = [*map(int,input().split())]
def getlis(arr):
    dp=[];lis=[]
    for num in arr:
        idx = bisect_left(dp,num)
        lis+=[idx+1]
        if len(dp)==idx:
            dp+=[num]
        else:
            dp[idx]=num
    return lis
lis = getlis(arr)
lds = getlis(arr[::-1])[::-1]
print(max(i+d-1 for i,d in zip(lis,lds)))
```
