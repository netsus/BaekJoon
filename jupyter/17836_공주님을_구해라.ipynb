### 문제


```python
! # 시작시간 체크
```

입력조건

    성의 크기 N행, M 열
    3 ≤ N, M ≤ 100
    공주 구하는 제한시간 T
    1 ≤ T ≤ 10000

```python
! # 끝 시간 체크
```

### 풀이

```python
from collections import deque
N, M, T = map(int,input().split())
castle=[list(map(int,input().split())) for _ in range(N)]
visited=[[0]*M for _ in range(N)]
dr=[1,-1,0,0]; dc=[0,0,1,-1]
sword=0; result=1e9

q=deque([(0,0)])
while q:
    r,c=q.popleft()
    if r==N-1 and c==M-1:
        result=visited[r][c]
        break
    for r_,c_ in zip(dr,dc):
        nr=r+r_;nc=c+c_
        if 0<=nr<N and 0<=nc<M and castle[nr][nc]!=1 and visited[nr][nc]==0:
            visited[nr][nc]=visited[r][c]+1
            q.append((nr,nc))
            if castle[nr][nc]==2:
                sword=nr,nc    
if sword:
    sr,sc=sword
    result = min(visited[sr][sc] + (N-1-sr) + (M-1-sc), result)
print(result if result<=T else "Fail")
```
