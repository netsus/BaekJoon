요약하자면, 틱택토 게임 9칸의 정보가 주어지면, 이게 valid한지 invalid한지 찾는 문제<br>
(첫번쨰는 반드시 X가 먼저 둔다.)

```python
! # 끝 시간 체크
```

### 풀이

```python
while 1:
    t_str=input()
    if t_str=="end": break
        
    x_n=t_str.count("X")
    o_n=t_str.count("O")

    case=[(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
    x_win=o_win=0
    for x,y,z in case:
        if t_str[x]==t_str[y]==t_str[z]:
            if t_str[x]=='X':
                x_win=1
            elif t_str[x]=='O':
                o_win=1
    valid=0
    if (x_win,o_win)==(1,0) and x_n==o_n+1:
        valid=1
    elif (x_win,o_win)==(0,1) and x_n==o_n:
        valid=1
    elif (x_win,o_win)==(0,0) and x_n==5 and o_n==4:
        valid=1
    print(['invalid','valid'][valid]) 
```

    상상코딩: 

    X가 먼저 두기 때문에 O와 개수가 같거나 1개 이상 많은 것만 valid 하다.
    X와 O의 개수가 같을 때는, O가 이긴 경우 이고,
    X가 1개 더 많은 경우는, X가 이겼거나, 비긴 경우 이다.
    그 외에는 모두 invalid 이다.
    핵심은 valid인 조건을 찾는 것이고, 그 외에는 invalid

    valid 경우의수
    X가 이긴경우
        X개수 = O개수+1
    O가 이긴경우
        X개수 = O개수
    비긴 경우
        X는 5개, O는 4개
    valid와 invalid를 나눈다.
