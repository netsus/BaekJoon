## 내풀이 합치기

```python
answer = 0 # 정답 : 도킹된 비행기 개수 
G = int(input()) # 게이트 개수
P = int(input()) # 비행기 개수

gate_parent = [i for i in range(G+1)] # 인덱스가 게이트 번호, 값이 대체 게이트번호

coming_plane=[] # 들어오는 비행기
for _ in range(P):
    coming_plane.append(int(input()))

def find(gate):
    if gate_parent[gate]==gate:
        return gate
    gate_parent[gate] = find(gate_parent[gate])
    return gate_parent[gate]

for plane in coming_plane:
    dck_gate = find(plane)
    if dck_gate==0:
        break
    gate_parent[dck_gate] = gate_parent[dck_gate-1] # 유니온 - 앞부분에 ㅇ녀결
    answer+=1

print(answer)
```

## 쪼개기

```python
def find_parent(x):
    if parent[x] == x:
        return x
    P = find_parent(parent[x])
    parent[x] = P
    return parent[x]

def union(x, y):
    x = find_parent(x)
    y = find_parent(y)
    if x < y:
        parent[y] = x
    else:
        parent[x] = y
```

```python
G = 4
P = 6

# union - find의 핵심
# {0:0, 1:1, 2:2, 3:3, 4:4}
parent = {i: i for i in range(0, G+1)}
parent
```

```python
planes = []
count = 0

for _ in [2,2,3,3,4,4]:
    planes.append(_) # planes.append(int(input().strip("\n")))
planes
```

### 핵심 알고리즘

```python
print(f"start parent {parent}")
for plane in planes:
    x = find_parent(plane)
    print(f"plane : {plane} - parent: {x}")
    if x == 0:
        break
    union(x, x-1)
    print(f"after union : {parent}")
    count += 1
```

## 전체 풀이 1

```python
import sys

def find_parent(x):
    if parent[x] == x:
        return x
    P = find_parent(parent[x])
    parent[x] = P
    return parent[x]

def union(x, y):
    x = find_parent(x)
    y = find_parent(y)
    if x < y:
        parent[y] = x
    else:
        parent[x] = y

G = int(input().strip("\n"))
P = int(input().strip("\n"))

# union - find의 핵심
# {0:0, 1:1, 2:2, 3:3, 4:4}
parent = {i: i for i in range(0, G+1)}
planes = []
count = 0

for _ in range(P):
    planes.append(int(input().strip("\n")))

for plane in planes:
    x = find_parent(plane)

    if x == 0:
        break
    union(x, x-1)
    count += 1

print(count)
```

## 전체코드 2

```python
# input = sys.stdin.readline
answer = 0
g = int(input())
p = int(input())
parent = [i for i in range(g + 1)]
planes = []

for _ in range(p):
    planes.append(int(input())) ## 비행기 들어오는 순서

def find(x): # 파인드함수
    if parent[x] == x: # 값이 자기자신인 경우 -> 루트인 경우
        return x # 루트를 반환 -> 과정에 있던 노드들에게 루트를 값으로 준다.
    parent[x] = find(parent[x]) # 재귀로 노드 한개 위로 올라가기
    return parent[x] # 최종 루트를 반환

for plane in planes:
    docking = find(plane) # 
    if docking == 0:
        break
    parent[docking] = parent[docking - 1] # docking
    answer += 1
print(answer)
```
