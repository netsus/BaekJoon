문제 요약

    화면에 A가 표시되어 있고, 버튼이 하나만 있는 기계가 있습니다.
    버튼을 누르면 A는 B로 변하고, B는 BA로 변하게 됩니다.
    K 번 버튼을 눌렀을 때, A와 B의 개수를 구해야 합니다.

```python
! # 끝 시간 체크
```

상상 코딩
    
    한 번 누르면 A:0 , B:1 입니다.
    그 다음은 A:1 B:1 -> A:1 B:2 -> A:2 B:3 입니다.
    A와 B가 피보나치의 규칙을 따르는 것을 볼 수 있습니다.

### 풀이

```python
K=int(input())
a,b=0,1
for i in range(1,K):
    a,b = b,a+b
print(a,b)
```

### 주석 달린 풀이

```python
K=int(input()) # 버튼을 누른 횟수 K
a,b=0,1 # 버튼을 1번 누른 경우로 초기화
for i in range(1,K): # 2번 누른것부터 for문 반복
    a,b = b,a+b # 피보나치 수열 진행
print(a,b) # A개수, B개수 출력
```
