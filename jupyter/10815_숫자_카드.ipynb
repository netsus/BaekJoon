문제요약
    
    첫 줄에 상근이가 가진 카드개수 N이 주어지고, 다음줄에 N개의 카드가 주어진다. 다음줄엔 구분해야할 카드 개수 M, 다음줄엔 M개의 카드가 주어진다.
    구분해야 할 카드들에 대해 상근이가 가진카드면 1, 아니면 0을 출력해야 합니다

상상코딩

    상근이 카드를 정렬하고, 구분해야할 카드가 들어오면 이분탐색해서 없으면 0, 있으면 1출력

```python
! # 끝 시간 체크
```

### 풀이

```python
import sys
input=sys.stdin.readline
N=int(input())
N_li=sorted(map(int,input().split()))
M=int(input())
M_li=list(map(int,input().split()))

result=[]
for m in M_li:
    s,e=0,N-1
    while s<=e:
        mid=(s+e)//2       
        if m==N_li[mid]:
            result.append(1)
            break
        elif m<N_li[mid]:
            e=mid-1
        else:
            s=mid+1
    if m!=N_li[mid]:
        result.append(0)
print(*result)
```

```python
print(*result)
```
