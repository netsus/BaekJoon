문제 요약
    
    게임 캐릭터 개수 N과 올릴 수 있는 레벨 총합 K가 주어진다.
    N명의 캐릭터에게 레벨 K를 골고루 올릴 때, 가능한 팀의 목표레벨(팀에서 제일 낮은 캐릭터의 레벨)중 최대 값을 출력

상상코딩

    최소레벨 ~ 최대레벨+K 사이를 이진탐색하며 가능한 최대 팀 목표레벨 T를 찾아갑니다.
    탐색하는 레벨을 t라고 했을 때 해당 레벨을 목표레벨로 만들기위해 팀원들이 레벨을 몇을 올려야 하는지 계산합니다.
    그 계산값이 K 이하이면 result에 저장하고 이분탐색을 하며 이 값을 갱신해갑니다.

```python
! # 끝 시간 체크
```

### 풀이

```python
import sys
input=sys.stdin.readline
N,K=map(int,input().split())
x=[int(input()) for _ in range(N)]

result=0
s,e=min(x),max(x)+K

while s <= e:
    trg_level=(s+e)//2
    need_level = sum([trg_level-l for l in x if trg_level-l >= 0])
    
    if need_level <= K:
        result = trg_level
        s = trg_level + 1
    elif need_level > K:
        e = trg_level - 1

print(result)
```
