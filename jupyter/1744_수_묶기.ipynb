```python
N=int(input())
l=[int(input()) for _ in range(N)]
positive_li=[]
negative_li=[]
result=0
for n in l:
    if n>1:
        positive_li.append(n)
    elif n==1: # 1  
        result+=1
    else: # 0  negative_li 
        negative_li.append(n)

positive_li.sort(reverse=True) # positive  
negative_li.sort() # negative  (   )

if len(positive_li)%2: # 1
    positive_li.append(1) # 1   
if len(negative_li)%2:
    negative_li.append(1)

for i in range(0,len(positive_li),2):
    result += positive_li[i]*positive_li[i+1]
for i in range(0,len(negative_li),2):
    result += negative_li[i]*negative_li[i+1]

print(result)
```
