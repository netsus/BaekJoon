    문제 요약

    수열이 주어졌을 때, 수열의 합을 구한다.
    다만, 수열 중 위치 상관없이 두 수를 묶을 수 있는데, 그 묶은 수는 서로 곱한 후 하나의 숫자로 취급한다.
    수열의 각 수를 적절히 묶었을 때, 그 합이 될 수 있는 최대값을 구하시오.
    ex) {0, 1, 2, 4, 3, 5} ⇒ 0 + 1 + (2*3) + (4*5) = 27
    ex) {-1, 0, 1} ⇒ (-1*0) + 1 = 1

```python
N=int(input())
l=[int(input()) for _ in range(N)]
positive_li=[]
negative_li=[]
result=0
for n in l:
    if n>1:
        positive_li.append(n)
    elif n==1: # 1은 무조건 더하기
        result+=1
    else: # 0 이하부터는 negative_li에 추가
        negative_li.append(n)

positive_li.sort(reverse=True) # positive는 큰것부터 곱하기
negative_li.sort() # negative는 작은것부터 곱하기(작은걸 곱하면 더 커진다)

if len(positive_li)%2: #홀수개라면 1추가
    positive_li.append(1) # 1은 곱해도 자기자신이기 때문
if len(negative_li)%2:
    negative_li.append(1)

for i in range(0,len(positive_li),2):
    result += positive_li[i]*positive_li[i+1]
for i in range(0,len(negative_li),2):
    result += negative_li[i]*negative_li[i+1]

print(result)
```
