### 풀이

```python
N = int(input())
need = [[1]*N for _ in range(N)]
result = 0
citymap = [list(map(int,input().split())) for _ in range(N)]

for k in range(N):
    for i in range(N):
        for j in range(N):
            if i==j or i==k or j==k:
                continue
            if citymap[i][j] == citymap[i][k] + citymap[k][j]:
                need[i][j] = False
            elif citymap[i][j] > citymap[i][k] + citymap[k][j]:
                result = -1
if result!=-1:
    for i in range(N):
        for j in range(i,N):
            if need[i][j]:
                result+=citymap[i][j]
print(result)
```
